```python
import numpy as np 
import pandas as pd 
import folium
from folium.plugins import Search
from functools import reduce
import math
# Convert BNG to LAT/LON
from bng_latlon import OSGB36toWGS84

df_nationwide_loanbook = pd.read_csv('data/output/NBS_Portfolio_Bristol_230216_csv/part-00000-70e38d50-6a64-46cb-b93c-ffc0af02033f-c000.csv')
```

```python
# Lambda function apply convertion of BNG to LAT/LON
vars_bng=['EASTING','NORTHING']
euk='EASTING'
nuk='NORTHING'
df_nationwide_loanbook['CORD'] = df_nationwide_loanbook[vars_bng].apply(lambda x : OSGB36toWGS84(x[euk],x[nuk]),axis=1)
df_nationwide_loanbook['LAT']=df_nationwide_loanbook['CORD'].apply(lambda p: p[0])
df_nationwide_loanbook['LON']=df_nationwide_loanbook['CORD'].apply(lambda p: p[1])
```

```python
# SAVE OUTPUT With Coordinates as CSV
# storing the data in JSON format
output_coord_csv_file='data/output/COORD_LATLON_NBS_Portfolio_Bristol_230216_csv/file_COORD_LATLON_NBS_Portfolio_Bristol_230216_csv.csv'
df_nationwide_loanbook.to_csv(output_coord_csv_file, header='true', sep=',', encoding='utf-8')
```

```python
# SAVE OUTPUT With Coordinates as JSON
# storing the data in JSON format
output_coord_json_file='data/output/COORD_LATLON_NBS_Portfolio_Bristol_230216_json/file_COORD_LATLON_NBS_Portfolio_Bristol_230216_json.json'
df_nationwide_loanbook.to_json(output_coord_json_file, orient = 'split', compression = 'infer', index = 'true')
```
