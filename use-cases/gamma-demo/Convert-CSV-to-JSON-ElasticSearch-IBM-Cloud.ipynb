```python
import pyspark
from pyspark.sql import functions as F
from pyspark.sql import SQLContext
from pyspark.sql import functions as F
```

```python
source_csv_path="data/raw_sligo_gamma_loanbook_csv/"
output_json_path='/notebooks-pv/logstash/'
sc = pyspark.SparkContext(appName="Convert CSV to JSON for Elastic Search Loading")
sqlContext = SQLContext(sc)
```

```python
internaldata_df1=sqlContext.read.csv(source_csv_path, header=True).persist(pyspark.StorageLevel.MEMORY_AND_DISK_2)
df1=internaldata_df1.filter("job_run_timestamp IS NOT NULL").filter("ecad_id IS NOT NULL").persist(pyspark.StorageLevel.MEMORY_AND_DISK_2)
```

```python
df2 = df1.drop("job_run_timestamp")
df3 = df2.withColumn('job_run_timestamp', F.current_timestamp())\
.withColumn('location', \
            F.struct( \
                     df2['etrs89_lat'].cast('float').alias('lat')  ,\
                     df2['etrs89_long'].cast('float').alias('lon') ))
```

```python
df4 = df3.select(df3['job_run_timestamp'],\
                 df3['ecad_id'].cast('integer'),\
                 df3['building_id'].cast('integer'),\
                 df3['address_line_1'].cast('string'),\
                 df3['address_line_2'].cast('string'),\
                 df3['address_line_3'].cast('string'),\
                 df3['address_line_4'].cast('string'),\
                 df3['address_line_5'].cast('string'),\
                 df3['address_line_6'].cast('string'),\
                 df3['address_line_7'].cast('string'),\
                 df3['address_line_8'].cast('string'),\
                 df3['address_line_9'].cast('string'),\
                 df3['etrs89_lat'].cast('float'),\
                 df3['etrs89_long'].cast('float'),\
                 df3['ber_rating'].cast('string'),\
                 df3['ber_rating_kwh'].cast('string'),\
                 df3['co2_emission'].cast('float'),\
                 df3['riverrp'].cast('float'),\
                 df3['rmax20'].cast('float'),\
                 df3['rmax75'].cast('float'),\
                 df3['rmax100'].cast('float'),\
                 df3['rmax200'].cast('float'),\
                 df3['rmax1000'].cast('float'),\
                 df3['sop_ri'].cast('float'),\
                 df3['coastaludrp'].cast('float'),\
                 df3['cudmax75'].cast('float'),\
                 df3['cudmax100'].cast('float'),\
                 df3['cudmax200'].cast('float'),\
                 df3['cudmax1000'].cast('float'),\
                 df3['swaterrp'].cast('float'),\
                 df3['swmax75'].cast('float'),\
                 df3['swmax200'].cast('float'),\
                 df3['swmax1000'].cast('float'),\
                 df3['model_river'].cast('string'),\
                 df3['model_coastal'].cast('string'),\
                 df3['model_sw'].cast('string'),\
                 df3['r20matrix'].cast('float'),\
                 df3['r75matrix'].cast('float'),\
                 df3['r100matrix'].cast('float'),\
                 df3['r200matrix'].cast('float'),\
                 df3['r1000matrix'].cast('float'),\
                 df3['cud75matrix'].cast('float'),\
                 df3['cud100matrix'].cast('float'),\
                 df3['cud200matrix'].cast('float'),\
                 df3['cud1000matrix'].cast('float'),\
                 df3['sw75matrix'].cast('float'),\
                 df3['sw200matrix'].cast('float'),\
                 df3['sw1000matrix'].cast('float'),\
                 df3['river_floodscore_ud'].cast('float'),\
                 df3['coastal_floodscore_ud'].cast('float'),\
                 df3['surfacewater_floodscore_ud'].cast('float'),\
                 df3['river_floodscore_def'].cast('float'),\
                 df3['floodscore_ud'].cast('float'),\
                 df3['floodscore_def'].cast('float'),\
                 df3['unflood_value'].cast('float'),\
                 df3['unflood_heightband'].cast('string'),\
                 df3['floodability_index_ud'].cast('string'),\
                 df3['floodability_index_def'].cast('string'),\
                 df3['Asset_id'].cast('integer'),\
                 df3['product'].cast('string'),\
                 df3['term'].cast('integer'),\
                 df3['Rate/Margin'].cast('float'),\
                 df3['Disbursment_Date'].cast('string'),\
                 df3['Closing_Date'].cast('string'),\
                 df3['Initial_Borrow_Capital'].cast('float'),\
                 df3['LTV'].cast('float'),\
                 df3['Monthly_Repayment'].cast('float'),\
                 df3['Next_Payment_Due'].cast('float'),\
                 df3['Monthly_Capital_Repayment'].cast('float'),\
                 df3['Months_Paid'].cast('integer'),\
                 df3['Months_Due'].cast('integer'),\
                 df3['Property_Valuation'].cast('float'),\
                 df3['Initial_Deposit'].cast('float'),\
                 df3['Mortgage_Outstanding_Balance'].cast('float') #,\
                 #df3['location']
                )
```

```python
#to place the file in the share path 
df4.write.mode("overwrite").format('json').save(output_json_path)
df4.printSchema()
print("Conversion CSV to JSON done!")
sc.stop()
exit()
```
